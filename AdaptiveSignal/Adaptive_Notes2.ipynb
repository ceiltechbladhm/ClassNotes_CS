# Lecture 2-27-2017

We talked about the block diagram of the steepest descent. We need to understand the weight parameters and the cost function $J$. We must define a weight-error vector. This is the error in the weight from the optimal value. Convergence properties of the steepest descent. We define a weighte rror vector $c(n)$:

$$ c(n) = w_o - w(n) $$

We want to get to the convergence of $w$. Fairly easy to see that as we approach the optimal value $w_o$, the error vector goes down. 

$$ c(n+1) = w_o - w(n+1) $$

We want to express the next weight error in terms of the current iteration.

$$ = w_o - \left[ w(n) + \mu(Rw_o - Rw(n)) \right] $$
$$ = c(n) - \mu Rc(n) $$
$$ =\left( I - \mu R \right)c(n) $$

Here, $I$ is the identity matrix. Expanding this in terms of the eigenvalue matrix $\Lambda$ and eigenvectors $Q$:

$$ I = \left( \mu Q \Lambda Q^H \right)c(n) $$

the goal is to study the errors as a function of the iteration. What is the time constant of $w$? To do this, we define a vector $c(n) = Q^H c(n)$. We are interested in a recursive operation again, so we look at the next iteration:

$$ v(n+1) = Q^H c(n+1) $$

Substitutiong our value for $c(n)$

$$ = \left(  Q^H - \mu Q^H Q \Lambda Q^H  \right)c(n) $$

$$ v(n+1) = \left( I - \mu \Lambda \right)v(n) $$

We can now easily study the trajectory of $v(n)$.

$$ v(n) = Q^H \left[ w_o - w(n) \right] $$

At $n=0$, $w(0) = 0$, $v(0) = Q^H w_o$. What is $v(1)$?

$$ v(1) = \left( I - \mu \Lambda \right)v(0) $$
$$ v(2) = \left( I - \mu \Lambda \right)^2 v(0) $$
$$ v(n) = \left( I - \mu \Lambda \right)^n v(0) $$

This is pretty nifty. The $kth$ element ($v_k(n)$), we can write:

$$ v_k(n) = \left( 1 - \mu\lambda_k \right)^n v_k(0) $$

For $k=1,...,M$. For stability, we need to make sure that $|1-\mu \lambda_k| < 1$. Therefore $-1 < 1 - \mu \lambda_k < 1$.

$$ -2 < - \mu \lambda_k < 0 $$
$$ 2 > \mu lambda_k > 0 $$

This means that $\mu > 0$ and $\mu < 2/\lambda_k$. This means that the upper bound on $\mu$ is given by $\mu_{max} = \dfrac{2}{\lambda_{max}}$. We define a time constant $\tau$ that determines the decaying exponential such that the max value drops by $1/e$.

Let's expand upon our equation: $v_k(n) = \left( 1 - \mu \lambda_k \right)^n v(0)$. If we expand this out, we can get:

$$ \left[ \left( 1 - \mu \lambda_k \right)^{-\tau} \right]^{-n\tau} $$

Since we know that we want the value to drop by $1/e$, we can figure out the value of $\tau$.

$$ e = \left( 1-\mu \lambda_k \right)^{-\tau} $$
$$ 1 = -\tau \ln\left( 1 - \mu \lambda_k \right) $$
$$ \tau_k = -\dfrac{1}{\ln\left(1-\mu\lambda_k\right)} $$

We can further approximate this if $\mu$ is small:

$$ \tau_k \approx \dfrac{1}{\mu \lambda_k} $$

We want to determine the slowest time constant.

$$w(n) = w_o - c(n)$$
$$ = w_o - Qv(n) $$

Here, $Q$ is $MxM$ matrix and $v(n)$ is $Mx1$. We can then write these in terms of a matrix:

$$ w(n) = w_o \left[\begin{matrix} q_1 & ... & q_m \end{matrix}\right]\left[\begin{matrix} v_1(n) \\ \vdots \\ v_m(n) \end{matrix}\right] $$
$$ = w_o - \sum_{k=1}^M q_k v_k (n) $$

Here, the sizes of each matrix is $w_o = Mx1$, $q_k = Mx1$ and $v_k(n) = 1x1$. Now if we take a certain element of $w$, and recall that we just solved for $v_k(n) = \left( 1 - \mu \lambda_k \right)^n v(0)$.

$$ w_i(n) = w_i - \sum_{k=1}^M q_{ki} v_k(n) $$




We can look at the convergence of error $J$.When we look at prediction, we see that we derivaed before that if we are not at an optimal wight, we will found ourself with the equation:

$$ J = J_{min} + \sum_{k=1}^M \lambda_k |v_k(n)|^2 $$

The important part, if we are to look at $|v_k(n)|^2 = \left[ \left( 1 - \mu \lambda_k \right)^2 \right]^n v_k(0)$. This quantity squared will be converging at a different time constant $\tau_{k,J}$. Roughly, without derivation and with small $\mu$, this will be 

$$ \tau_{k,J} \approx \dfrac{1}{2\mu \lambda_k} = \dfrac{1}{2} \tau_{k,w} $$

This is saying that the error time constant will converge half as quickly as the weight parameters (I think???).
