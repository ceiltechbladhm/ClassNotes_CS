# Regularization

We have a model given by $\hat{m} = A_{true}f + \epsilon$

The goal is to find some $A$ which approximates $A_{true}$.

The matrix A comes from an ill-posed inverse problem. Variational regularization. The idea is to take the argument (defined an argumend $f$ which was defined in space $R^n$). The idea is to minimize the following operation: $||Af - m||^2_2$. The trouble with ill-posed inverse problems is that there are a large number of $f$ which give us very small numbers for this operation. This is why we need regularization.

We add another penalty. One from error; another from apriori information. We use $\alpha$. We use this to emphasize which term we value more.

The starting point:



\begin{array}{cc|c}
  & \alpha ||f||_2^2 & |f_1|^2 + |f_2|^2 + ... + |f_n|^2 \\
 ||Af - m||^2_2 + & \alpha ||Lf||_2^2 & |f_2 - f_1|^2 + |f_3-f_2|^2 + ... + |f_n-f_{n-1}|^2 \\
 & \alpha ||Lf||_1 & |f_2-f_1| + |f_3 - f_2| + ... + |f_n - f_{n-1}|
\end{array}

Note the last one is the Euclidian 1 norm.

Here, $L$ is a finite difference matrix. 

$$ 
L = \left[\begin{matrix} -1 & +1 & 0 & 0 & ...\\
  0 & -1 & + 1 & 0 & ... \\
   &  & ... &  &\end{matrix}\right]
$$

We will only be focusing on the 1 norm (the last option).



Matlab's quadprog:
argmin (x) : $\dfrac{1}{2} x^T Hx + h^T x$ such that there is $Bx \leq b$, and also $C_x = c$ (equality constraint), and also $lowerbound \leq x_j \leq upperbound$. The function to be minimied has a second order term, and a linear term.
