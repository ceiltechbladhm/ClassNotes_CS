ALM/CLB - basically collection of LUTs and bits. Have registers, and chains, The outputs could come directly out of LUTs. They also allow you to directly tap into the register. Local interconnect (intra-CLB connections). Internally, they also have some feedback paths from the outputs. This will be very fast and cheap.

Some level of flexibility from routing and the CLB. Same is true of the output. In general terms, this is called connection box. Connection box flexibility says what percentave from the innerconnect can make connections to the local logic. Full connectivity means any wire can connect to any of the inputs. Full flexibility means no constraints on tools downstream. Also lots and lots of switches (6 sram switches each). 

Other is called the switch box (SB). The tracks make up the large portion of area used. SB has its own flexibility.

Fracturable LUTs. Altera used efficiently.

inputs have some delay for LUTs. Potentially optimize the bit which changes most often by using the input with the least delay.
