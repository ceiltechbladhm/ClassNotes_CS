What is a necessary condition for a solution to be optimal locally. For a given solution, we can imply that these conditions are true:

$ \nabla f(x^*) = 0 $ (first order condition)

One thing to clarify: first we started with a 1D problem. IN this case, we prove necessary condition. We write down definition. We can verifythat it must be equal to zero. Next, we moved to the case where we have vector as a variable. We clarify:

$f(x^*)$  - local minimum

We construct a scalar condition $\alpha$ for any fixed $d$.

$$ g(\alpha) = f(x^* + \alpha d) $$

If $x^*$ is a local min for a problem, it implies that $\alpha=0$ is a local min for min $g(\alpha)$ for any given $d$.

Then we can use the fact that $0$ is a local min $g(\alpha) \implies $ condition 1 is true. 

It is homework to prove that. If we look at every direction and is the smallest, must be local min. 

$ f(x^*) $ local min

$\implies$ there exists $\epsilon > 0$ such that $f(x^* + c) \geq f(x^*)$ for all $||v|| < \epsilon$

For any $d$, $\alpha \leq \dfrac{\epsilon}{||d||}$

$$ f(x^* + \alpha d) \geq f(x^*) $$ for all $\alpha \leq \dfrac{\epsilon}{||d||}$

$$ \implies g(\alpha) \geq g(0) $$ for all $\alpha \leq \dfrac{\epsilon}{||d||}$



for very simple problem: (w is known)

$$ \min(x,y) \dfrac{1}{2}(x\cdot y - w)^2 $$

Question: can we find all stationary solutions ($\geq 0$), and all local global min

How can we apply optimality conditions to have some understanding about this problem?

$$ f^\prime(x) = 0 $$
$$ f^{\prime \prime} \geq 0 $$

Let's get to it.

$$ \dfrac{\partial f(x,y)}{\partial x} = (xy - w)y = xy^2 - wy $$
$$ \dfrac{\partial f(x,y)}{\partial y} = x(xy - w) = x^2 - xw $$

$$ \dfrac{\partial f}{\partial x} \implies \left[\begin{matrix} y = 0 \\ xy = w \end{matrix}\right] $$
$$ \dfrac{\partial f}{\partial y} \implies \left[\begin{matrix} x = 0 \\ xy = w \end{matrix}\right] $$

These two cannot happen at the same time. Two conclusions:

1. (0,0)
2. ($x$, $\dfrac{w}{x}$)

These are stationary solutions: they satisfy the first order condition. Other solutions I don't need to consider because they do not meet the first condition. Among these, we can say a little bit more. We can see that (0,0) isn't a good solution. Can we prove it? In order to do this, we need the second order. We need Haschon matrix???

$$ \left[\begin{matrix} y^2 & 2xy - w\\ 2xy - w & x^2 \end{matrix}\right] $$

Let's first try H(0,0). We get:

$$ \left[\begin{matrix} 0 & -w \\ -w & 0 \end{matrix}\right] \geq 0 $$

This is not true (as long as $w$ is positive). Let's verify

$$ \left[\begin{matrix} x & y \end{matrix}\right] \left[\begin{matrix} 0 & -w \\ -w & 0 \end{matrix}\right] \left[\begin{matrix} x \\ y \end{matrix}\right] \implies -2xyw \geq 0 $$

This means that (0,0) is not a local min. ??? Don't understand why exactly.

Now try $H(x,\dfrac{w}{x}$:

$$ H(x,\dfrac{w}{x} = \left[\begin{matrix} \dfrac{w^2}{x^2} & w \\ w & x^2 \end{matrix}\right] $$

Try this. Multiply ab left and right (ab are arbitrary substitutions for xy)

$$ \left[\begin{matrix} a & b \end{matrix}\right] \left[\begin{matrix} \dfrac{w^2}{x^2} & w \\ w & x^2 \end{matrix}\right] \left[\begin{matrix} a \\ b \end{matrix}\right] $$

$$ = \dfrac{w^2}{x^2}a^2 + x^2 b^2 + 2abw $$
$$ = (\dfrac{wa}{x} + xb)^2 \geq 0 $$

This is true because of the square! It turns out that all local mins are global mins for this particular problem. We haven't proved this yet.

* $f(a) - f(b) = f^\prime(c)(a-b)$

Assume that these are true: let's prove the sufficient condition.

$$ f(x) - f(x^*) = f^\prime(x^*)(x - x^*) + \dfrac{1}{2}(x-x^*) ^2 f^{\prime \prime}(x^* + \alpha(x - x^*) $$

We can express anything inbetween for any $\alpha \exist [0,1]$ through $x^* + \alpha(x - x^*)$. We know to be true from mean value theorem.

Apply the sufficient condition for any $x$ sufficiently close to $x^*$:

$$ f(x) - f(x^*) = \dfrac{1}{2}(x - x^*)^2 f^{\prime \prime}(x^* + \alpha(x - x^*)) \geq 0 $$

Because of the square., $f^{\prime \prime}(x^* + \alpha(x - x^*)) \geq 0$.  Because $^{\prime \prime}(x^*) \geq 0$, there must exist a neighborhood close enough such that $f^{\prime \prime}(x) \geq 0$. No matter how small a region I take, there must be a region which is curving up (because of local min). Recall the meaning between second derivative.

$$ A = \left[\begin{matrix} a_1^T\\a_2^T \\ ... \\ a_N^T \end{matrix}\right] $$
