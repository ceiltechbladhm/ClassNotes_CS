What is an estimator? it is a function  $g$ of random data. $x[n]$ is random data. 

$$ \hat{\theta} = g\left( x[0], x[1], ... ,x[N-1] \right) $$

If there is nothing random, there is no estimation. There are things I don't know; are these things random or deterministic?

Parameter vs waveform estimation. Parameter estimation pertains to a parameter $\theta$ whose dimension is fixed and does not depend on $N$. Waveform estimation depends on the number of elements collected. We will talk about both. 

* Estimate - value returned by an estimator
* Estimator - function $g$ which returns an estimate
* Estimation - process of figuring out the estimator

We need to clarify how the randomness manifests itself.

$$ p(x[0]; \theta) = \dfrac{1}{\sqrt{2\pi \sigma^2}} e^{-\dfrac{1}{2\sigma^2} \left( x[0] - \theta \right)^2} ~N(\theta,\sigma^2) $$

In other cases, without the probability density funciton ,we have a generator model. This is the data which generates it. We can try to fit a straight line:

$$ x[n] = \theta_t + \theta_2 n +w[n] $$

Where $w[n] ~ N(0,\sigma^2)$. The mean of $x$ is given by $\theta_{1,2}$. we can multiply and figure out from the model:

$$ p(x;\theta) = \dfrac{1}{\sqrt{2\pi \sigma^2}} e^{-\dfrac{1}{2\sigma^2} \sum_{n=1}^{N-1} \left(x[n] - \theta_1 - \theta_2n \right)^2} $$

Parameters are viewed as random (Bayes framework) $p(x,\theta) = p(x|\theta)p(\theta)$ or deterministic $p(x;\theta)$. Pay attention  to something! The semicolon indicates that $\theta$ is deterministic; the comma shows that it is random. We will show it to capture a probability function. If someone tells us that $\theta$ falls within a range, it may be better to use a probability density function within the range.

## Chapter 2 Minimum-variance unbiased Estimation

Deterministic $\theta$: Bias $b(\theta) := E\left[ \hat{\theta} \right] - \theta$. Unbiased is : $E\left[ \hat{\theta} = \theta \forall \theta \right]$

Example 1:

$$ x[n] = A+ w[n] $$

1. $\hat{A} = \dfrac{1}{N} \sum_{n=0}^{N-1} x[n] \implies E\left( \hat{A} \right) = A$
2. $\hat{A} = \dfrac{1}{2N} \sum_{n=0}^{N-1} x[n] \implies E\left( \dfrac{1}{2}\hat{A} \right) = A \begin{cases} = A, \text{ if } A=0\\ \neq A, \text{ if } A=0 \end{cases}$

It may be a good idea, to ahve several estimators, if they are uncorrelated. If we combin estimator,s we can maintain biases. We can reduce the variance.

$$ \hat{\theta} = \dfrac{1}{n} \sum_{i=1}^n \hat{\theta}_i $$

Reasonable performance metric? The mean square error has a bias term in it. 

$$ mse(\hat{\theta}) = E\left[ \hat{\theta} - \theta \right]^2 =\text{var}(\hat{\theta}) + b^2(\theta) $$

MSE estimators may not be realizable. In a random variable

Biased estimator and unbiased estimator. Unbiasedness may depend on the parameter itself. If we claim it is biased or not biased. The mean is parameter dependent.

Consider an estimator such that the mean squared error is minimized.

$$ mse(\hat{A}) = \dfrac{\alpha^2 \sigma^2}{N} + \left( \alpha-1 \right)^2 A^2 $$

We find that the optimum alpha is dependent on the known parameter A. Do this by setting derivative of MSE=0, and solve for parameter $\alpha$.

$$ \dfrac{\partial MSE}{\partial \alpha} = 0 $$
$$ \alpha_{opt} = \dfrac{A^2}{A^2 + \sigma^2/N} $$

It needs to minimize the variance of the estimator. What we want to understand is that if we try to combine vias andvariance, your may be unrealizable. 

Questipon: Does MVUE (Minimum variance unbiased estimation) always exist? Not in general
Question: Can we always find it? Not in general.
MVUEs may not exist.

The dependence on the value of the estimator depends on the value itself. It may be impossible to find an unbiased estimator. 

If an MVUE esists, the Cramer-Rao lower bound will tell you what it exists.
Rao-Blackwell-Lehmann-Scheffe Thm on sufficient statistics will allow us to identify MVUE.
If all else fails, restrict the estimators to be linear.
