## Physical disturbances

If density changes other physical properties change too.

$\rho(x,t) = \rho_o + \rho_1(x,t)$
$$p(x,t) = p_o + p_1(x,t)$$
$$T(x,t) = T_o + T_1(x,t)$$

Here:
* $\rho$ = density
* $p$ = pressure
* $T$ = Temperature

In a small space, each space has a given pressure, temterature, and density ($p,T, \rho$ respectively)

When a pressure wave occurs, temperature will oscillate as well. Is usually small.

When pressure wave comes in, density varies along the coordinate. At wavefront, density is increased. Behind, density is increased. In front of wavefront, no change. Because of the difference in pressure, there is an oscillatory flow of molecules. Thus, one more physical quantity changes: fluid velocity $v$.

$$v(x,t) = 0 + v_1(x,t)$$

Thus, the quantities which represent the acoustic wave are:

* $\rho_1(\hat{x},t)$
* $p_1(\hat{x},t)$
* $T_1(\hat{x},t)$
* $v_1(\hat{x},t)$

## Relationship between quantities

We want to derive the relationship between quantities

Here:
* $V$ - volume
* $K$ - adiabatic compressibility (no heat)
* $S$ - entropy

### Relationship between Density and Pressure
$$ K = -\dfrac{1}{V}\dfrac{\partial V}{\partial P} = \dfrac{1}{\rho}\dfrac{\partial \rho}{\partial P}$$

Note all of these are made for a given entropy

$$\Delta S = 1/T \Delta Q = 0$$

When $K$ is finite, the fluid is compressible. Solving $K$,

$$ \rho_1 = K \rho_o \rho_1$$

### Relationship between Density and Temperture
To determine elevation of temperture, hold the following condition true:

$$\Delta Q = 0$$

Also note the thermal expansion coeff, $\beta$.

$$ \beta =  \dfrac{1}{V}\dfrac{\partial V}{\partial T}$$
$$ \gamma = \dfrac{c_p}{c_v} $$
* $dq_1$ - change in pressure
* $dq_2$ - change in temperture
* Require $dq_1$ and $dq_2$ to sum to 0

$$ dq_1 = T dx = T\dfrac{\partial S}{\partial p} dp = -T \dfrac{\partial V}{\partial T}dp$$
$$ dq_2 = C_pdT $$

Begin with $$\Delta Q = 0$$

$$ dq_2 = dq_1 $$
$$ C_pdT = T\dfrac{\partial V}{\partial T} dp $$
$$ \Delta T = \dfrac{TV\beta \Delta p}{C_p} $$
$$ \Delta T = \Delta p (\gamma-1)TK $$
